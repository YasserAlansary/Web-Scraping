```python
url='https://www.forbes.com/self-made-women/'
driver_path= 'msedgedriver'

browser= Edge(executable_path= driver_path)
browser.get(url)
```

```python
Names=[]
```

```python
for i in range(4):
    try:
        names = browser.find_elements(By.XPATH,'//div[starts-with(@class,"table-row")]')

        for name in names:
            txt = name.text
            Names.append(txt)
        
        print(Names)
        Next = browser.find_element(By.XPATH,'//span[text()="Next "]').click()
    
    except ElementClickInterceptedException:
        print('The End\n')
```

```python
Names
```

```python
Data =[]
for i in range(len(Names)):
    name = Names[i]
    data = name.split('\n')
    Data.append(data)
    
Data
```

```python
Rank = []
Name = []
Net_Worth = []
Age = []
Source = []
State = []
```

```python
for i in range(len(Data)):
    value = Data[i]
    for i in range(len(value)):
        v = value[i]
        
        if i == 0:
            Rank.append(v)
        elif i == 1:
            Name.append(v)
        elif i == 2:
            Net_Worth.append(v)
        elif i == 3:
            Age.append(v)
        elif i == 4:
            Source.append(v)
        elif i == 5:
            State.append(v)
        elif i == 6:
            Rank.append(v)
        else:
            pass
```

```python
Name
```

```python
# Top 10 richest Women in USA 
for i in range(10):
    print(Rank[i], Name[i], Net_Worth[i], Source[i], State[i])
```
