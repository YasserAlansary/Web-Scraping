```python
from selenium.webdriver import Edge
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.common.by import By
from selenium.webdriver import ActionChains
from selenium.webdriver.support.ui import Select
from selenium.webdriver.support import expected_conditions
from selenium.common.exceptions import NoSuchElementException
from selenium.common import exceptions  
from openpyxl import load_workbook
import pandas as pd 
import numpy as np
```

```python
url='https://companiesmarketcap.com/'
driver_path= 'msedgedriver'

browser= Edge(executable_path= driver_path)
browser.get(url)
```

```python

category = browser.find_elements(By.XPATH,'//span[starts-with(@class,"option")]')

categories=[]
for cat in category:
    text= cat.text
    categories.append(text)
    
del categories[-1]
categories
```

```python
categories.insert(7,'More +')
```

```python
del categories[7]
categories
```

```python
writer = pd.ExcelWriter('F:\\New folder\\Companies.xlsx', engine='xlsxwriter')
i = 0

for cat in categories:
    if i < 8:
        element_tab= browser.find_element(By.XPATH, f'//span[text()="{cat}"]').click()
        element = browser.find_elements(By.XPATH,'//table[starts-with(@class,"default-table table marketcap-table dataTable")]')
        df = pd.read_html(browser.page_source)[0]
        df.to_excel(writer, sheet_name = f'sheet{i}', index=False)
        print(df)
    else:
        
        element_tab= browser.find_element(By.XPATH, f'//span[text()="{cat}"]').click()
        element = browser.find_elements(By.XPATH,'//table[starts-with(@class,"default-table table marketcap-table dataTable")]')
        df = pd.read_html(browser.page_source)[0]
        df.to_excel(writer, sheet_name = f'sheet{i}', index=False)
        print(df)
        element_tab= browser.find_element(By.XPATH, f'//span[text()="More +"]').click()
        
    i+=1
    
writer.save()
```
