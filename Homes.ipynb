```python
from selenium.webdriver import Edge
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.common.by import By
from selenium.webdriver import ActionChains
from selenium.webdriver.support.ui import Select
from selenium.webdriver.support import expected_conditions
from selenium.common.exceptions import NoSuchElementException
from selenium.common.exceptions import InvalidSelectorException, NoSuchElementException, ElementNotInteractableException
from selenium.common.exceptions import StaleElementReferenceException
import pandas as pd 
import numpy as np
import time 
```

```python
url='https://www.homes.com/'
driver_path= 'msedgedriver'

browser= Edge(executable_path= driver_path)
browser.get(url)
```

```python
searsh= browser.find_element(By.XPATH, '//*[@id="ta-ge2k"]')

searsh.clear()
searsh.send_keys('New York')
searsh.send_keys(Keys.ENTER)
```

```python
text=[]
```

```python
for i in range(5):
    try:
        features = browser.find_elements(By.XPATH,'//div[starts-with(@class,"for-sale-content-container")]')
        for i in range(len(features)):
            try:
                txt= features[i].text
                text.append(txt)
                #time.sleep(5)
                
            except(StaleElementReferenceException):
                txtt= features[i].get_attribute("innerHTML")
                text.append(txtt)
                #time.sleep(5)
            
    except(InvalidSelectorException ,StaleElementReferenceException ,ElementNotInteractableException):
        text.append('NA')
        
    Next= browser.find_element(By.XPATH,'//button[starts-with(@class,"next text-only")]').click()
    
```

```python
features = browser.find_elements(By.XPATH,'//div[starts-with(@class,"for-sale-content")]')

for i in range(len(features)):
    txt= features[i].text
    Text.append(txt)
    
Next= browser.find_element(By.XPATH,'//button[starts-with(@class,"next text-only")]').click()

len(Text)
```

```python
Texts= set(Text)
len(Texts)
```

```python
data= []
```

```python
for i in range(len(Text)):
    apartment= Text[i]
    apart= apartment.split("\n")
    
    data.append(apart)
```

```python
Houses= pd.DataFrame(data)
Houses.to_excel('Houses.xlsx')
```

```python
price= []
bed= []
bath= []
Ft= []
address= []
description= []
agent= []
```

```python
for j in range(len(data)):
    
    dep= data[j]
    for i in range(len(dep)): 
        value= dep[i]
        if i== 0:
            if '$' in value:
                price.append(value)
            else: 
                price.append('NA')
        elif i== 1:
            if 'Bed' in value:
                bed.append(value)
            else: 
                bed.append('NA')
        elif i== 2:
            if 'Bath' in value:
                bath.append(value)
            else: 
                bath.append('NA')
        elif i== 3:
            if 'Sq Ft' in value:
                Ft.append(value)
            else: 
                Ft.append('NA')
        elif i== 4:
            if 'New York' in value:
                address.append(value)
            else: 
                address.append('NA')
        elif i== 5:
            if value.isspace():
                description.append('NA')  
            else:
                description.append(value)
        elif i== 6:
            if value.isspace():
                agent.append('NA')  
            else:
                agent.append(value)
        else:
            pass
                        
print(price,address)
```

```python
des= len(price)- len(description)
agt= len(price)- len(agent)

for i in range(des):
    description.append('NA')

for i in range(agt):
    agent.append('NA')
```

```python
Data= pd.DataFrame()

Data['price']= price 
Data['bed']= bed
Data['bath']= bath
Data['Ft']= Ft
Data['address']= address
Data['description']= description
Data['agent']= agent

Data
```
