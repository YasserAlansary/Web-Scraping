```python
Detailed = browser.find_element(By.XPATH, '//*[text()="Detailed"]').click()
```

```python
names = []
Names = []
Positions = []

elements = browser.find_elements(By.XPATH,'//table[starts-with(@class,"inline-table")]')

for i in range(len(elements)):
    element = elements[i].text
    if element == "":
        pass
    else:
        names.append(element)

for i in range(len(names)):
    value = names[i]
    name = value.split('\n')
    for i in range(len(name)):
        if i == 0:
            Names.append(name[i])
        elif i == 1:
            Positions.append(name[i])
    
print(Names, "\n\n", Positions)
```

```python
ages= []

elements = browser.find_elements(By.XPATH,'//td[starts-with(@class,"zentriert")]')

for i in range(len(elements)):
    element = elements[i].text
    if element == "":
        pass
    else:
        ages.append(element)
len(ages)
```

```python
Rank= []
Age= []
Matchs= []
Goals= []
Assists= []
Yellow_Cards= []
Red_Cards =[]
Substitutions_On = []
Substitutions_Offs= []

n = 11
data = [ages[i:i + n] for i in range(0, len(ages), n)]

for i in range(len(data)):
    values = data[i]
    for j in range(len(values)):
        
        value = values[j]
        if j == 0:
            Rank.append(value)
        elif j == 1:
            Age.append(value)
        elif j == 2:
            Matchs.append(value)
        elif j == 3:
            Goals.append(value)
        elif j == 5:
            Assists.append(value)
        elif j == 6:
            Yellow_Cards.append(value)
        elif j == 8:
            Red_Cards.append(value)
        elif j == 9:
            Substitutions_On.append(value)
        elif j == 10:
            Substitutions_Offs.append(value)
        else:
            pass
```

```python
data= pd.DataFrame()

data['Rank']= Rank
data['Name']= Names
data['Position']= Positions
data['Age']= Age
data['Matchs']= Matchs
data['Goals']= Goals
data['Assists']= Assists
data['Yellow_Cards']= Yellow_Cards
data['Red_Cards']= Red_Cards
data['Substitutions_On']= Substitutions_On
data['Substitutions_Offs']= Substitutions_Offs

data
```
