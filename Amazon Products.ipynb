```python
from selenium.webdriver import Edge
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.common.by import By
from selenium.webdriver import ActionChains
from selenium.webdriver.support.ui import Select
from selenium.webdriver.support import expected_conditions
from selenium.common.exceptions import NoSuchElementException
from selenium.common.exceptions import InvalidSelectorException, ElementNotInteractableException 
from openpyxl import load_workbook
import pandas as pd 
import numpy as np
```

```python
url='https://www.amazon.com/'
driver_path= 'msedgedriver'

browser= Edge(executable_path= driver_path)
browser.get(url)
```

```python
search_field_id = 'twotabsearchtextbox'
element_searsh_field = browser.find_element(By.ID, search_field_id)

element_searsh_field.clear()
element_searsh_field.send_keys('Dresses')
element_searsh_field.send_keys(Keys.ENTER)
```

```python
#Up to 3 stars
stars = browser.find_element(By.XPATH,'//i[starts-with(@class,"a-icon a-icon-star-medium a-star-medium-3")]')
actions= ActionChains(browser)
actions.move_to_element(stars)
actions.click(stars)
actions.perform()

#actions.doubleClick(stars).perform()
```

```python

"""
names = browser.find_elements(By.XPATH,'//span[starts-with(@class,"a-size-base-plus a-color-base a-text-normal")]')
    
for i in range(len(names)):
    name = names[i].text
    if name == "":
        pass
    else:
        Names.append(name)
#Names
"""

for page in range(3):
    Items= []
    names = browser.find_elements(By.XPATH,'//span[starts-with(@class,"a-size-base-plus")]')
    
    for i in range(len(names)):
        name = names[i].text
        Items.append(name)
        NAMES.append(name)
    
    print(len(Names))
    
    for i in range(len(Names)):
        try:

            Name = Items[i]
            name = browser.find_element(By.XPATH, f'//*[text()="{Name}"]').click()

            try:   
                star = browser.find_element(By.XPATH,'//span[starts-with(@class,"a-icon-alt")]')
                Star = star.get_attribute("innerHTML")
                Stars.append(Star)
            except (InvalidSelectorException ,NoSuchElementException ,ElementNotInteractableException):
                star = browser.find_element(By.XPATH,'//i[starts-with(@class,"a-icon a-icon-star ")]')
                Star = star.get_attribute("innerHTML")
                Stars.append(Star)
                
                #Stars.append('NA')

            try:
                review = browser.find_element(By.XPATH,'//*[@id="acrCustomerReviewText"]')
                Review = review.text
                Reviews.append(Review)
            except (InvalidSelectorException ,NoSuchElementException ,ElementNotInteractableException):
                Reviews.append('NA')

            try:
                price = browser.find_element(By.XPATH,'//span[starts-with(@class,"a-offscreen")]')
                Price = price.get_attribute("innerHTML")
                Prices.append(Price)
            except (InvalidSelectorException ,NoSuchElementException ,ElementNotInteractableException):
                Prices.append('NA')
                
            try:
                feature = browser.find_element(By.XPATH,'//*[@id="feature-bullets"]')
                Feature = feature.text
                Features.append(Feature)
            except (InvalidSelectorException ,NoSuchElementException ,ElementNotInteractableException):
                Features.append('NA')

            browser.back()
            print(Star, Review,Price)

        except (InvalidSelectorException ,NoSuchElementException ,ElementNotInteractableException):
            Prices.append('NA')
            Stars.append('NA')
            Reviews.append('NA')
            Features.append('NA')
    
    
    browser.back()
    browser.refresh()
    browser.refresh()
    Next = browser.find_element(By.XPATH, '//*[text()="Next"]').click()
    #Next_Page()

```
