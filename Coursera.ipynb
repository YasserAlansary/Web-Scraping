```python
url='https://www.coursera.org/'
driver_path= 'msedgedriver'

browser= Edge(executable_path= driver_path)
browser.get(url)
```

```python
element_searsh_field = browser.find_element(By.XPATH, '//input[starts-with(@class,"react-autosuggest")]')

element_searsh_field.clear()
element_searsh_field.send_keys('Data Science')
element_searsh_field.send_keys(Keys.ENTER)
```

```python
Xpath='//*[@id="cds-react-aria-46"]'
```

```python
Names = []
Stars = []
Reviews = []
Levels = []
Reviews = []
Skills = []
common=['University of Colorado Boulder', 'IBM','','Credit Eligible','CertNexus',
       'University of Michigan','University of Illinois','Illinois Institute of Technology'
       ,'Ball State University','University of Leeds','IIT Roorkee','University of Illinois at Urbana-Champaign']
```

```python
for i in range(11):
    names = browser.find_elements(By.XPATH,'//h3[starts-with(@class,"cds-33 cds-CommonCard-title css-1wot210 cds-35")]')
    stars = browser.find_elements(By.XPATH,'//p[starts-with(@class,"cds-33 css-zl0kzj cds-35")]')
    levels = browser.find_elements(By.XPATH,'//p[starts-with(@class,"cds-33 css-14d8ngk")]')
    reviews = browser.find_elements(By.XPATH,'//p[starts-with(@class,"cds-33 css-14d8ngk cds-35")]')

    for i in range(len(names)):
        name = names[i].text
        if name == '':
            pass
        else:
            Names.append(name)

    for i in range(len(stars)):
        star = stars[i].text
        if star is None:
            Stars.append('null')
        else:    
            Stars.append(star)

    for i in range(len(levels)):
        level = levels[i].text
        if 'Skills' in level or 'reviews' in level or level in common:
            pass 
        else:
            Levels.append(level)     

    for i in range(len(levels)):
        level = levels[i].text
        if 'Skills' in level:
            Skills.append(level) 
        else:
            pass

    for i in range(len(reviews)):
        review = reviews[i].text
        if 'reviews' in review:
            Reviews.append(review)
        else:
            pass


    Pages = browser.find_element(By.XPATH,Xpath).click()

print(len(Names), "\n", len(Stars),'\n',len(Levels),len(Reviews), len(Skills))
```

```python
missed = ['58',77,124,125,128,129]

for i in range(len(missed)):
    pos = int(missed[i])
    Stars.insert(pos, 'unknown')
    Reviews.insert(pos, 'unknown')
    if pos == 125:
        Skills.insert(pos, 'unknown')
    else:
        pass
    
    
print(len(Names), "\n", len(Stars),'\n',len(Levels),len(Reviews), len(Skills))
```

```python
Level = []
Type = []
Period = []

for i in range(len(Levels)):
    txt = []
    txt = Levels[i].split("·", 2)
    Level.append(txt[0])
    Type.append(txt[1])
    Period.append(txt[2])
    
print(Level[100],Type[100],Period[100])
```
